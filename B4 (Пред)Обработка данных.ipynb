```python
import pandas as pd
import numpy as np

log = pd.read_csv('log.xls', header=None)
log.columns = ['user_id','time','bet','win']

def cleaner(user_id):
    if user_id == '#error':
        return log.drop(np.where(log['user_id'] == '#error')[0])
    else:
        index_id = user_id.find('user')
        user_id = user_id[index_id::]
        return user_id

log.user_id = log.user_id.apply(cleaner)
log
```

```python
import pandas as pd
import numpy as np

log = pd.read_csv('log.xls', header=None)
log.columns = ['user_id','time','bet','win']

log = log.drop(np.where(log['user_id'] == '#error')[0])
log['time']=log['time'].str.replace('[', '')

def normolise_user_id(user_id):
    index_id = user_id.find('user')
    user_id = user_id[index_id::]
    return user_id
    
log.user_id = log.user_id.apply(normolise_user_id)

log
```
